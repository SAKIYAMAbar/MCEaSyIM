#**Diffusion Rule**

##**Multi Campaign Independent Cascade model (MCICM) [1]**
Two types of information (Correct information and Misinformation) propagate on the network: corrective information and erroneous information.

This program assumes that the diffusion of misinformation occurs at time
$t = 0$, followed by that of correct information at time $t = r$.

Rules are as follows:

1. This program starts a diffusion process of misinformation at time $t = 0$ with a set of 10 seed nodes  selected uniformly at random, $A_\text{M}$.
2. This program compute the set $A_\text{C}$ (Initial seed node set to spread correct information) using a given
algorithm at time $t = r$ when the fraction of nodes activated with the misinformation is 0.05 or larger
for the first time. Then, we start a diffusion process of correct information with the seed set $A_\text{C}$ at time $t = r$.

•$p_\text{M}(u, v) \in (0, 1)$ : In misinformation, the probability that an
active node $u$ activates their inactive out-neighbor $v$.

•$p_\text{C}(u, v) \in (0, 1)$ : In correct information, the probability that an
active node $u$ activates their inactive out-neighbor $v$.

• In terms of Multi Campaign Independent Cascade model (MCIC) model, $p_\text{M}(u, v)=0.1$, $p_\text{C}(u, v)=0.05$.

• In terms of Multi Campaign Weighted Cascade model (MCWC) model, $p_\text{M}(u, v)=d_v$, $p_\text{C}(u, v)=1/2d_v$.   ($d_v$ is in-degree of node $v$)
